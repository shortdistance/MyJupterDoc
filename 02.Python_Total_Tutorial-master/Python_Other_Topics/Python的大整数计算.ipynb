# Python的大整数计算

python中的大数整数四则运算非常的快,而且不限大小(当然了还是受限于内存大小的),这让我很奇怪,于是查了下,原来python的Long类型其实是不固定长度的,类似向量.因此才会如此省事好用.其实像c像java中也有这样的包,只是不是默认支持需要调用第三方包而已.

```python
from random import randint
```

```python
a = randint(1e307,1e308)
```

```python
a
```

```python
b = randint(1e307,1e308)
```

```python
b
```

```python
a*b
```

```python
%timeit a*b
```

```python
import operator
```

```python
reduce(operator.mul,[i for i in range(1,1001)])#1000!
```

```python
%timeit reduce(operator.mul,[i for i in range(1,100001)])
```

是不是很黄很暴力!
